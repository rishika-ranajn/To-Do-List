```python
import pandas as pd
import numpy as np

# Create empty DataFrame
tasks_df = pd.DataFrame(columns=["Task", "Status"])


def show_tasks():
    if tasks_df.empty:
        print("\nNo tasks available.")
    else:
        print("\nYour Tasks:")
        print(tasks_df)


def add_task():
    global tasks_df
    task = input("Enter new task: ")
    
    new_row = pd.DataFrame({
        "Task": [task],
        "Status": ["Pending"]
    })
    
    tasks_df = pd.concat([tasks_df, new_row], ignore_index=True)
    print("Task added!")


def remove_task():
    global tasks_df
    show_tasks()
    
    try:
        index = int(input("Enter task index to remove: "))
        tasks_df = tasks_df.drop(index)
        tasks_df = tasks_df.reset_index(drop=True)
        print("Task removed!")
    except:
        print("Invalid index!")


def mark_done():
    global tasks_df
    show_tasks()
    
    try:
        index = int(input("Enter task index to mark done: "))
        tasks_df.loc[index, "Status"] = "Done"
        print("Task marked as done!")
    except:
        print("Invalid index!")


def stats():
    total = len(tasks_df)
    done = np.sum(tasks_df["Status"] == "Done")
    pending = total - done
    
    print("\n📊 Task Stats:")
    print("Total:", total)
    print("Done:", done)
    print("Pending:", pending)


def menu():
    print("\n===== TO-DO LIST (PANDAS) =====")
    print("1. View Tasks")
    print("2. Add Task")
    print("3. Remove Task")
    print("4. Mark Done")
    print("5. Show Stats")
    print("6. Exit")


# Main loop
while True:
    menu()
    choice = input("Enter choice: ")

    if choice == '1':
        show_tasks()

    elif choice == '2':
        add_task()

    elif choice == '3':
        remove_task()

    elif choice == '4':
        mark_done()

    elif choice == '5':
        stats()

    elif choice == '6':
        print("Exiting... 👋")
        break

    else:
        print("Invalid choice!")
```
